```python
import pandas as pd
import requests

API_KEY = 'AIzaSyAbQSFXSdcpZ-yY_F-uya-lJQjlwirdhfM'

def get_place_details(latitude, longitude):
    url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={latitude},{longitude}&radius=50&type=restaurant&key={API_KEY}"
    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()

        if 'results' in data and data['results']:
            place_id = data['results'][0]['place_id']
            details_url = f"https://maps.googleapis.com/maps/api/place/details/json?place_id={place_id}&fields=name,formatted_address,delivery,dine_in,curbside_pickup,reservable,takeout,wheelchair_accessible_entrance,reviews,rating,price_level&key={API_KEY}"
            details_response = requests.get(details_url)

            if details_response.status_code == 200:
                details_data = details_response.json()
                restaurant_name = details_data['result'].get('name', 'NA')
                address = details_data['result'].get('formatted_address', 'NA')
                delivery = details_data['result'].get('delivery', 'NA')
                dine_in = details_data['result'].get('dine_in', 'NA')
                curbside_pickup = details_data['result'].get('curbside_pickup', 'NA')
                reservable = details_data['result'].get('reservable', 'NA')
                takeout = details_data['result'].get('takeout', 'NA')
                wheelchair_accessible_entrance = details_data['result'].get('wheelchair_accessible_entrance', 'NA')
                rating = details_data['result'].get('rating', 'NA')
                price_level = details_data['result'].get('price_level', 'NA')
                num_reviews = details_data['result'].get('user_ratings_total', 'NA')
                types = details_data['result'].get('types', 'NA')

                reviews = []
                if 'reviews' in details_data['result']:
                    reviews = [review['text'] for review in details_data['result']['reviews']]

                return {
                    'restaurant_name': restaurant_name,
                    'address': address,
                    'delivery': delivery,
                    'dine_in': dine_in,
                    'curbside_pickup': curbside_pickup,
                    'reservable': reservable,
                    'takeout': takeout,
                    'wheelchair_accessible_entrance': wheelchair_accessible_entrance,
                    'rating': rating,
                    'price_level': price_level,
                    'reviews': reviews,
                    'num_reviews':num_reviews,
                    'types':types
                }

    nearby_places_url = f"https://maps.googleapis.com/maps/api/place/nearbysearch/json?location={latitude},{longitude}&radius=500&type=restaurant&key={API_KEY}"
    nearby_places_response = requests.get(nearby_places_url)

    if nearby_places_response.status_code == 200:
        nearby_places_data = nearby_places_response.json()
        if 'results' in nearby_places_data and nearby_places_data['results']:
            nearby_place_latitude = nearby_places_data['results'][0]['geometry']['location']['lat']
            nearby_place_longitude = nearby_places_data['results'][0]['geometry']['location']['lng']
            return get_place_details(nearby_place_latitude, nearby_place_longitude)

    return {
        'restaurant_name': 'NA',
        'address': 'NA',
        'delivery': 'NA',
        'dine_in': 'NA',
        'curbside_pickup': 'NA',
        'reservable': 'NA',
        'takeout': 'NA',
        'wheelchair_accessible_entrance': 'NA',
        'rating': 'NA',
        'price_level': 'NA',
        'reviews': [],
        'num_reviews':'NA',
        'types':'NA'
    }

csv_file = 'combined_data.csv'
data = pd.read_csv(csv_file)

restaurant_names = []
addresses = []
delivery_lst = []
dine_in_lst = []
curbside_pickup_lst = []
reservable_lst = []
takeout_lst = []
wheelchair_accessible_entrance_lst = []
rating_lst = []
price_level_lst = []
reviews_lst = []
num_reviews=[]
types=[]

for index, row in data.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    place_info = get_place_details(latitude, longitude)
    restaurant_names.append(place_info['restaurant_name'])
    addresses.append(place_info['address'])
    delivery_lst.append(place_info['delivery'])
    dine_in_lst.append(place_info['dine_in'])
    curbside_pickup_lst.append(place_info['curbside_pickup'])
    reservable_lst.append(place_info['reservable'])
    takeout_lst.append(place_info['takeout'])
    wheelchair_accessible_entrance_lst.append(place_info['wheelchair_accessible_entrance'])
    rating_lst.append(place_info['rating'])
    price_level_lst.append(place_info['price_level'])
    reviews_lst.append(place_info['reviews'])
    num_reviews.append(place_info['num_reviews'])
    types.append(place_info['types'])

data['restaurant_name'] = restaurant_names
data['address'] = addresses
data['delivery'] = delivery_lst
data['dine_in'] = dine_in_lst
data['curbside_pickup'] = curbside_pickup_lst
data['reservable'] = reservable_lst
data['takeout'] = takeout_lst
data['wheelchair_accessible_entrance'] = wheelchair_accessible_entrance_lst
data['rating'] = rating_lst
data['price_level'] = price_level_lst
data['reviews'] = reviews_lst
data['num_reviews']=num_reviews
data['types']=types

# Write the DataFrame back to the CSV file
data.to_csv('combined_file.csv', index=False)
```

```python
df.drop(columns=['id','address_1','address_2','address_3','state_province',
                 'postal_code','phone','website_url','street','restaurant_name','reviews',
                 'num_reviews','types'], inplace=True)
```

```python
def remove_outliers(df, column):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

df = remove_outliers(df, 'rating')
```

```python
df['rating'] = df['rating'].astype(str)
df2 = df.drop(df[df['rating'].str.strip() == ''].index)
```

```python
columns_to_convert = ['rating', 'delivery', 'dine_in', 'curbside_pickup','reservable','takeout','wheelchair_accessible_entrance']
df2[columns_to_convert] = df2[columns_to_convert].apply(pd.to_numeric)
df2[columns_to_convert] = df2[columns_to_convert].replace({'TRUE': 1, 'FALSE': 0})
df2[columns_to_convert] = df2[columns_to_convert].fillna(-1)
```

```python
df2['price_level'] = df2['price_level'].fillna(-1)
```

##Visulizations

### How does the average brewery rating vary across different states in the United States?

```python
breweries = df[df['country'] == 'United States']
plt.figure(figsize=(10, 10))
ax = plt.subplot(111, polar=True)
num_states = len(breweries['state'].unique())

theta = np.linspace(0, 2*np.pi, num_states, endpoint=False).tolist()
average_ratings = breweries.groupby('state')['rating'].mean().tolist()
ax.plot(theta, average_ratings, color='purple', linewidth=2)

ax.set_xticks(theta)
ax.set_xticklabels(breweries['state'].unique(), fontsize=6)

plt.title('Average Brewery Ratings Across US States', fontsize=14)
plt.show()
```

The polar plot visualizes the average brewery ratings across different states in the United States. Each state is represented by a point on the circumference of the plot, with the distance from the center indicating the average rating of breweries in that state. The plot reveals variations in brewery ratings across states, with some states having higher average ratings compared to others. For instance, states with points farther from the center tend to have higher average ratings, while those closer to the center have lower average ratings. This visualization allows for quick comparisons of brewery ratings across states and highlights regions with particularly high or low-rated breweries.
